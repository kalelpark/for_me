# 가우시안 블러, 노이즈를 제거 하며, 이미지를 흐리게함

- 커널 사이즈 변화에 따른 흐림
- 표준 편차 변화에 따른 흐림

```python
import cv2
img = cv2.imread('./pirpo.png')

# 커널 사이즈는 양수의 홀수를 사용합니다. (대체적으로 (3, 3), (5, 5), (7, 7))

kernel_3 = cv2.GaussianBlur(img, (3, 3), 0)  # 커널 사이즈, 자동으로 표준편차 선택 : 0
kernel_5 = cv2.GaussianBlur(img, (5, 5), 0)
kernel_7 = cv2.GaussianBlur(img, (7, 7), 0)


cv2.imshow('img', kernel_3)
cv2.imshow('img', kernel_5)
cv2.imshow('img', kernel_7)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 이미지 변형(원근)

```python
import cv2
import numpy as np

img = cv2.imread('./newspaper.jpg')

width, height = 640, 240  # 가로 크기 640, 세로크기 240 으로 결과물 출력

# 4개의 점소환  (시계방향으로 4지점 정의\)
src = np.array([[511, 352], [1008, 345], [1122, 584], [455, 594]], dtype = np.float32)  # input 지점
dst = np.array([[0, 0], [width, 0], [width, height], [0 ,height]], dtype = np.float32)  # output ㅣㅈ점


# src -> dst
matrix = cv2.getPerspectiveTransform(src, dst)  # matrix 생성

# 적용
result = cv2.warpPerspective(img, matrix, (width, height))  # matrix 변환

cv2.imshow('img', img)
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 이미지 펼치기

```python
import cv2
import numpy as np

img = cv2.imread('./poker.jpg')

width, height = 530, 710  

# 4개의 점소환  (시계방향으로 4지점 정의\)
src = np.array([[702, 143], [1133, 414], [726, 1007], [276, 700]], dtype = np.float32)  # input 지점
dst = np.array([[0, 0], [width, 0], [width, height], [0 ,height]], dtype = np.float32)  # output ㅣㅈ점


# src -> dst
matrix = cv2.getPerspectiveTransform(src, dst)  # matrix 생성

# 적용
result = cv2.warpPerspective(img, matrix, (width, height))  # matrix 변환


cv2.imshow('img', img)
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 마우스 이벤트 등록

```python
import cv2

def mouse_handler(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:    # LB 버튼 누름
        print('왼쪽 버튼 Down')
        print(x, y)   
    elif event == cv2.EVENT_LBUTTONUP: # 왼쩍 마우스를 때는 경우
        print('왼쪽 버튼  UP')
        print(x, y) 
    elif event == cv2.EVENT_LBUTTONDBLCLK: # 더블클릭
        print('더블클릭')
    elif event == cv2.EVENT_MOUSEMOVE:  # 마우스 이동
        print('마우스 이동')
    
img = cv2.imread('./poker.jpg')
cv2.namedWindow('img')    # Img라는 이름의 윈도를 생성하는 것
                          # 추후 마우스 이벤트를 처리하기 위한 핸들러 적용
    
cv2.setMouseCallback('img', mouse_handler)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 미니 프로젝트

```python
import cv2
import numpy as np

point_list = []


def mouse_handler(event, x, y, flags, param):
    global drawing 
    dst_img = src_img.copy()
    
    if event == cv2.EVENT_LBUTTONDOWN:    
        drawing = True
        point_list.append((x, y))
        
    if drawing:   
        prev_point = None  # 직선의 시작점
        for point in point_list:
            cv2.circle(dst_img, point, 15, COLOR, cv2.FILLED)
            if prev_point:
                cv2.line(dst_img, prev_point, point, COLOR, THICKNESS, cv2.LINE_AA)
            prev_point = point
            
        next_point = (x, y)
        if len(point_list) == 4:
            show_result()    # 결과 출력
            next_point = point_list[0]
            
        cv2.line(dst_img, prev_point, next_point, COLOR, THICKNESS, cv2.LINE_AA)

    
    
    cv2.imshow('img', dst_img)

def show_result():
    width, height = 530, 710  
    
    src = np.float32(point_list)
    dst = np.array([[0, 0], [width, 0], [width, height], [0 ,height]], dtype = np.float32)
    
    matrix = cv2.getPerspectiveTransform(src, dst)
    result = cv2.warpPerspective(src_img, matrix, (width, height))
    cv2.imshow('result', result)
    
    
src_img = cv2.imread('./newspaper.jpg')
cv2.namedWindow('img')
COLOR = (255, 0, 255)
THICKNESS = 3
drawing = False  #  선 그릴지 여부 선택

cv2.setMouseCallback('img', mouse_handler)
cv2.imshow('img', src_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
